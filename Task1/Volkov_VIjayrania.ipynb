```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

df1 - dataframe with gender=1<br/>
df0 - dataframe with gender=0

```python
df = pd.read_csv(filepath_or_buffer='heightC.txt', sep=' ', skiprows=2)
df1 = df.loc[df["gender"] == 1]
df0 = df.loc[df["gender"] == 0]
```

Let's calculate the mean weight of both genders and visualize them

```python
m0 = df0["weight"].mean()
m1 = df1["weight"].mean()
sns.barplot(y=[m0, m1], x=[0, 1])
```

Now let's do the same with the mean height

```python
m0 = df0["height"].mean()
m1 = df1["height"].mean()
sns.barplot(y=[m0, m1], x=[0, 1])
```

Based on the fact, that gender 0 species are in average significantly heavier and slightly taller, we can assume that gender 0 is male, and gender 1 is female

```python
df0_x = np.arange(len(df0))
df1_x = np.arange(len(df1))

fig, ((ax00, ax01))  = plt.subplots(1,2, figsize=(14,8), sharey= True)

ax00.plot( df0_x, 'height', data=df0, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4,  label="Individual")
ax00.plot( df0_x, 'momHeight', data=df0, marker='', color='olive', linewidth=2,  label="Mom")
ax00.plot( df0_x, 'dadHeight', data=df0, marker='', color='olive', linewidth=2, linestyle='dashed', label="Dad")
ax00.set_ylabel('height')
ax00.set_title("gender 0 height with respective parent's height")
ax00.legend()

ax01.plot( df1_x, 'height', data=df1, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4,  label="Individual")
ax01.plot( df1_x, 'momHeight', data=df1, marker='', color='olive', linewidth=2,  label="Mom")
ax01.plot( df1_x, 'dadHeight', data=df1, marker='', color='olive', linewidth=2, linestyle='dashed', label="Dad")
ax01.set_ylabel('height')
ax01.set_title("gender 1 height with respective parent's height")
ax01.legend()

fig, (ax10, ax11) = plt.subplots(1,2, figsize=(14,8), sharey= True)
ax10.plot( df0_x, 'weight', data=df0, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4,  label="Individual")
ax10.plot( df0_x, 'momWeight', data=df0, marker='', color='olive', linewidth=2,  label="Mom")
ax10.plot( df0_x, 'dadWeight', data=df0, marker='', color='olive', linewidth=2, linestyle='dashed', label="Dad")
ax10.set_ylabel('weight')
ax10.set_title("gender 0 weight with respective parent's weight")
ax10.legend()

ax11.plot( df1_x, 'weight', data=df1, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4,  label="Individual")
ax11.plot( df1_x, 'momWeight', data=df1, marker='', color='olive', linewidth=2,  label="Mom")
ax11.plot( df1_x, 'dadWeight', data=df1, marker='', color='olive', linewidth=2, linestyle='dashed', label="Dad")
ax11.set_ylabel('weight')
ax11.set_title("gender 1 weight with respective parent's weight")
ax11.legend
```

The plot 1st in first row above shows that gender 0's height is closer to Dad's height than mom's height
The plot 2nd in first row above shows that gender 1's height is closer to Mom's height than Dad's height

The plot 3rd above shows that gender 0's weight is not related to parent's weight and is highly variable.
The plot 4th above shows that gender 1's weight is closer to mom's weight than Dad's weight .

# Overall, we find out that gender 0 is showing more characteristics of Dad and gender 1 is shwoing more characteristics of Mom so it is safe to say that gender 0 denotes 'Male' and gender 1 denotes 'female' in the data.
